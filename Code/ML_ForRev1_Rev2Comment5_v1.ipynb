```python
y = [0.8, 0.7, 0.4]
x = [0, 1, 2]

age_labels = ['Yo', 'MA', 'El']
x_ym = [0, 1]
y_ym = [y[0], y[1]]  # top line

x_mo = [1, 2]
y_mo = [y[1], y[2]]  # top line

x_yo = [0, 2]
y_yo = [y[0], y[2]]  # top line

plt.figure(figsize=(6, 4))

# Plot lines
plt.plot(x_ym, y_ym, marker='o', color='red', markerfacecolor='black', markeredgecolor='black', linewidth=1.5, label = 'Yo-MA')
plt.plot(x_mo, y_mo, marker='o', color='blue', markerfacecolor='black', markeredgecolor='black', linewidth=1.5, label = 'MA-El')
plt.plot(x_yo, y_yo, marker='o', color='green', markerfacecolor='black', markeredgecolor='black', linewidth=1.5, label = 'Yo-El')

# Vertical dashed lines
for i in range(len(x)):
    plt.plot([x[i], x[i]], [0, y[i]], linestyle='dashed', color='black', linewidth=1)

# Vertical dashed lines
for i in range(len(x)):
    plt.plot([-0.2, x[i]], [y[i], y[i]], linestyle='dashed', color='black', linewidth=1)

# Formatting
plt.xticks(ticks=[0, 1, 2], labels=age_labels)
plt.yticks([0.1, 0.9], labels=['small', 'large'])
plt.ylabel('Predicted Mean Well-Being')
plt.xlabel('Age')
plt.ylim(0, 1)
plt.xlim(-0.2, 2.2)
plt.grid(False)
plt.legend()
plt.tight_layout()

plt.show()
```

### Plot together

```python
import matplotlib.pyplot as plt

# Define the 9 well-being patterns
y_list = [
    [0.8, 0.7, 0.4],
    [0.8, 0.8, 0.4],
    [0.8, 0.4, 0.4],
    [0.4, 0.7, 0.8],
    [0.4, 0.7, 0.7],
    [0.4, 0.4, 0.7],
    [0.4, 0.7, 0.45],
    [0.75, 0.4, 0.7],
    [0.6, 0.6, 0.6]
]

Title_list = ['a: Younger Enjoy Case 1', 'b: Younger Enjoy Case 2', 'c: Younger Enjoy Case 3',
             'd: Older Enjoy Case 1', 'e: Older Enjoy Case 2', 'f: Older Enjoy Case 3',
             'g: Middle-age Enjoy', 'h:Middle-age Suffer', 'i: No Trend']

age_labels = ['Yo', 'MA', 'El']
x = [0, 1, 2]

fig, axes = plt.subplots(3, 3, figsize=(14, 12))
axes = axes.flatten()

for idx, y in enumerate(y_list):
    ax = axes[idx]
    # Define line segments
    x_ym, y_ym = [0, 1], [y[0], y[1]]
    x_mo, y_mo = [1, 2], [y[1], y[2]]
    x_yo, y_yo = [0, 2], [y[0], y[2]]

    # Plot lines with black markers
    ax.plot(x_ym, y_ym, marker='o', color='red', markerfacecolor='black', markeredgecolor='black', linewidth=1.5, label='Yo-MA')
    ax.plot(x_mo, y_mo, marker='o', color='blue', markerfacecolor='black', markeredgecolor='black', linewidth=1.5, label='MA-El')
    ax.plot(x_yo, y_yo, marker='o', color='green', markerfacecolor='black', markeredgecolor='black', linewidth=1.5, label='Yo-El')

    ax.annotate('', xy=(x_ym[1], y_ym[1]), xytext=(x_ym[0], y_ym[0]),
            arrowprops=dict(arrowstyle='->,head_length=.75,head_width=.5', color='red'))
    ax.annotate('', xy=(x_mo[1], y_mo[1]), xytext=(x_mo[0], y_mo[0]),
                arrowprops=dict(arrowstyle='->,head_length=.75,head_width=.5', color='blue'))
    ax.annotate('', xy=(x_yo[1], y_yo[1]), xytext=(x_yo[0], y_yo[0]),
                arrowprops=dict(arrowstyle='->,head_length=.75,head_width=.5', color='green'))

    # Vertical dashed lines
    for i in range(len(x)):
        ax.plot([x[i], x[i]], [0, y[i]], linestyle='dashed', color='black', linewidth=1)
        ax.plot([-0.2, x[i]], [y[i], y[i]], linestyle='dashed', color='black', linewidth=1)

    # Formatting
    ax.set_xticks(ticks=[0, 1, 2])
    ax.set_xticklabels(age_labels)
    ax.set_yticks([0.1, 0.9])
    ax.set_yticklabels(['small', 'large'])
    ax.set_ylim(0, 1)
    ax.set_xlim(-0.2, 2.2)
    ax.grid(False)
    ax.set_title(f'{Title_list[idx]}', loc = 'left')

# Shared labels
fig.text(0.5, 0.04, 'Age', ha='center', fontsize=12)
fig.text(0.04, 0.5, 'Predicted Mean Well-Being', va='center', rotation='vertical', fontsize=12)

# Optional: add legend to first subplot only
axes[0].legend(loc='upper right')

plt.tight_layout(rect=[0.05, 0.05, 1, 0.95])

plt.savefig('Figures/Fig5relationship.jpg', format='jpg', dpi=300, bbox_inches='tight')

plt.show()
```
