## Import Package

```python
import os 
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

## Load Data

```python
DoubleWashedDataset_Filename = os.path.join("Data", "GallupWB_Zhang73var525k6wave_v1.parquet")
```

```python
DoubleWashedDataset = pd.read_parquet(DoubleWashedDataset_Filename)
```

```python
DoubleWashedDataset.shape
```

## Train and Test df

```python
DoubleWashedDataset = DoubleWashedDataset.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
DoubleWashedDataset.head()
```

```python
y = DoubleWashedDataset['Cantril_ladder']
```

```python
X = DoubleWashedDataset.drop(columns=['Cantril_ladder'])
```

```python
X.shape
```

```python
X_encoded = pd.get_dummies(X, columns=["COUNTRY_ISO3"])
```

```python
X_train_encoded, X_test_encoded, y_train, y_test = train_test_split(X_encoded, y, test_size=0.1, random_state=42)
```

## Analysis Test

### Random forest

#### TEST_1

```python
rf_model = RandomForestRegressor(n_estimators=500, random_state=42, max_depth = 10, oob_score=True, n_jobs=-1)
rf_model.fit(X_train_encoded, y_train)
```

```python
y_pred = rf_model.predict(X_test_encoded)
```

```python
r2_score(y_test, y_pred) * 100
```

```python
y_pred = rf_model.predict(X_train_encoded)
```

```python
r2_score(y_train, y_pred) * 100
```
