```python
wave_level_df = []

for wave in set(wave_df['wave']):
    print(f"Wave: {wave}")
    YoungTreatedAsYoung_wave_select = YoungTreatedAsYoung_wave[YoungTreatedAsYoung_wave['wave']==wave]
    YoungTreatedAsMiddle_wave_select = YoungTreatedAsMiddle_wave[YoungTreatedAsMiddle_wave['wave']==wave]
    YoungTreatedAsOld_wave_select = YoungTreatedAsOld_wave[YoungTreatedAsOld_wave['wave']==wave]
    MiddleTreatedAsYoung_wave_select = MiddleTreatedAsYoung_wave[MiddleTreatedAsYoung_wave['wave']==wave]
    MiddleTreatedAsMiddle_wave_select = MiddleTreatedAsMiddle_wave[MiddleTreatedAsMiddle_wave['wave']==wave]
    MiddleTreatedAsOld_wave_select = MiddleTreatedAsOld_wave[MiddleTreatedAsOld_wave['wave']==wave]
    OldTreatedAsYoung_wave_select = OldTreatedAsYoung_wave[OldTreatedAsYoung_wave['wave']==wave]
    OldTreatedAsMiddle_wave_select = OldTreatedAsMiddle_wave[OldTreatedAsMiddle_wave['wave']==wave]
    OldTreatedAsOld_wave_select = OldTreatedAsOld_wave[OldTreatedAsOld_wave['wave']==wave]
    line = [wave]
    
    YoungYmTreatmentDifference = YoungTreatedAsYoung_wave_select['Predict_y'].mean() - YoungTreatedAsMiddle_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(YoungTreatedAsYoung_wave_select['Predict_y'], YoungTreatedAsMiddle_wave_select['Predict_y'])
    print(f"Wave: {wave} Young Treatment Difference (Y - M): {YoungYmTreatmentDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [YoungYmTreatmentDifference, t_stat, p_value]
    
    YoungYoTreatmentDifference = YoungTreatedAsYoung_wave_select['Predict_y'].mean() - YoungTreatedAsOld_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(YoungTreatedAsYoung_wave_select['Predict_y'], YoungTreatedAsOld_wave_select['Predict_y'])
    print(f"Wave: {wave} Young Treatment Difference (Y - O): {YoungYoTreatmentDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [YoungYoTreatmentDifference, t_stat, p_value]

    YoungMoTreatmentDifference = YoungTreatedAsMiddle_wave_select['Predict_y'].mean() - YoungTreatedAsOld_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(YoungTreatedAsMiddle_wave_select['Predict_y'], YoungTreatedAsOld_wave_select['Predict_y'])
    print(f"Wave: {wave} Young Treatment Difference (M - O): {YoungMoTreatmentDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [YoungMoTreatmentDifference, t_stat, p_value]
    
    MiddleYmTreatmentDifference = MiddleTreatedAsYoung_wave_select['Predict_y'].mean() - MiddleTreatedAsMiddle_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(MiddleTreatedAsYoung_wave_select['Predict_y'], MiddleTreatedAsMiddle_wave_select['Predict_y'])
    print(f"Wave: {wave} Middle Treatment Difference (Y - M): {MiddleYmTreatmentDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [MiddleYmTreatmentDifference, t_stat, p_value]

    MiddleYoTreatmentDifference = MiddleTreatedAsYoung_wave_select['Predict_y'].mean() - MiddleTreatedAsOld_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(MiddleTreatedAsYoung_wave_select['Predict_y'], MiddleTreatedAsOld_wave_select['Predict_y'])
    print(f"Wave: {wave} Middle Treatment Difference (Y - O): {MiddleYoTreatmentDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [MiddleYoTreatmentDifference, t_stat, p_value]

    MiddleMoTreatmentDifference = MiddleTreatedAsMiddle_wave_select['Predict_y'].mean() - MiddleTreatedAsOld_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(MiddleTreatedAsMiddle_wave_select['Predict_y'], MiddleTreatedAsOld_wave_select['Predict_y'])
    print(f"Wave: {wave} Middle Treatment Difference (M - O): {MiddleMoTreatmentDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [MiddleMoTreatmentDifference, t_stat, p_value]

    OldYmTreatmentDifference = OldTreatedAsYoung_wave_select['Predict_y'].mean() - OldTreatedAsMiddle_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(OldTreatedAsYoung_wave_select['Predict_y'], OldTreatedAsMiddle_wave_select['Predict_y'])
    print(f"Wave: {wave} Old Treatment Difference (Y - O): {OldYmTreatmentDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [OldYmTreatmentDifference, t_stat, p_value]
    
    OldYoTreatmentDifference = OldTreatedAsYoung_wave_select['Predict_y'].mean() - OldTreatedAsOld_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(OldTreatedAsYoung_wave_select['Predict_y'], OldTreatedAsOld_wave_select['Predict_y'])
    print(f"Wave: {wave} Old Treatment Difference (Y - O): {OldYoTreatmentDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [OldYoTreatmentDifference, t_stat, p_value]

    OldMoTreatmentDifference = OldTreatedAsMiddle_wave_select['Predict_y'].mean() - OldTreatedAsOld_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(OldTreatedAsMiddle_wave_select['Predict_y'], OldTreatedAsOld_wave_select['Predict_y'])
    print(f"Wave: {wave} Old Treatment Difference (M - O): {OldMoTreatmentDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [OldMoTreatmentDifference, t_stat, p_value]
    print('---------------------')
    
    YoungYmTreatedAgeDifference = YoungTreatedAsYoung_wave_select['Predict_y'].mean() - MiddleTreatedAsYoung_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(YoungTreatedAsYoung_wave_select['Predict_y'], MiddleTreatedAsYoung_wave_select['Predict_y'])
    print(f"Wave: {wave} Young Treated Age Difference (F - M): {YoungYmTreatedAgeDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [YoungYmTreatedAgeDifference, t_stat, p_value]
    
    YoungYoTreatedAgeDifference = YoungTreatedAsYoung_wave_select['Predict_y'].mean() - OldTreatedAsYoung_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(YoungTreatedAsYoung_wave_select['Predict_y'], OldTreatedAsYoung_wave_select['Predict_y'])
    print(f"Wave: {wave} Young Treated Age Difference (F - O): {YoungYoTreatedAgeDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [YoungYoTreatedAgeDifference, t_stat, p_value]
    
    YoungMoTreatedAgeDifference = MiddleTreatedAsYoung_wave_select['Predict_y'].mean() - OldTreatedAsYoung_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(MiddleTreatedAsYoung_wave_select['Predict_y'], OldTreatedAsYoung_wave_select['Predict_y'])
    print(f"Wave: {wave} Young Treated Age Difference (M - O): {YoungMoTreatedAgeDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [YoungMoTreatedAgeDifference, t_stat, p_value]

    MiddleYmTreatedAgeDifference = YoungTreatedAsMiddle_wave_select['Predict_y'].mean() - MiddleTreatedAsMiddle_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(YoungTreatedAsMiddle_wave_select['Predict_y'], MiddleTreatedAsMiddle_wave_select['Predict_y'])
    print(f"Wave: {wave} Middle Treated Age Difference (F - M): {MiddleYmTreatedAgeDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [MiddleYmTreatedAgeDifference, t_stat, p_value]
    
    MiddleYoTreatedAgeDifference = YoungTreatedAsMiddle_wave_select['Predict_y'].mean() - OldTreatedAsMiddle_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(YoungTreatedAsMiddle_wave_select['Predict_y'], OldTreatedAsMiddle_wave_select['Predict_y'])
    print(f"Wave: {wave} Middle Treated Age Difference (F - O): {MiddleYoTreatedAgeDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [MiddleYoTreatedAgeDifference, t_stat, p_value]
    
    MiddleMoTreatedAgeDifference = MiddleTreatedAsMiddle_wave_select['Predict_y'].mean() - OldTreatedAsMiddle_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(MiddleTreatedAsMiddle_wave_select['Predict_y'], OldTreatedAsMiddle_wave_select['Predict_y'])
    print(f"Wave: {wave} Middle Treated Age Difference (M - O): {MiddleMoTreatedAgeDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [MiddleMoTreatedAgeDifference, t_stat, p_value]    
    
    OldYmTreatedAgeDifference = YoungTreatedAsOld_wave_select['Predict_y'].mean() - MiddleTreatedAsOld_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(YoungTreatedAsOld_wave_select['Predict_y'], MiddleTreatedAsOld_wave_select['Predict_y'])
    print(f"Wave: {wave} Old Treated Age Difference (F - M): {OldYmTreatedAgeDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [MiddleYmTreatedAgeDifference, t_stat, p_value]
    
    OldYoTreatedAgeDifference = YoungTreatedAsOld_wave_select['Predict_y'].mean() - OldTreatedAsOld_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(YoungTreatedAsOld_wave_select['Predict_y'], OldTreatedAsOld_wave_select['Predict_y'])
    print(f"Wave: {wave} Middle Treated Age Difference (F - O): {OldYoTreatedAgeDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [MiddleYoTreatedAgeDifference, t_stat, p_value]
    
    OldMoTreatedAgeDifference = MiddleTreatedAsOld_wave_select['Predict_y'].mean() - OldTreatedAsOld_wave_select['Predict_y'].mean()
    t_stat, p_value = stats.ttest_ind(MiddleTreatedAsOld_wave_select['Predict_y'], OldTreatedAsOld_wave_select['Predict_y'])
    print(f"Wave: {wave} Old Treated Age Difference (M - O): {OldMoTreatedAgeDifference:.4f} T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    line = line + [OldMoTreatedAgeDifference, t_stat, p_value]    
    
    print('------------------\n')
    wave_level_df.append(line)
```
